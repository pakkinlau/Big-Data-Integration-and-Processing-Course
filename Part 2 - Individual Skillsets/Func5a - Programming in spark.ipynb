Spark in High level perspective:

- Top: Separate functionalities that Spark provides
- Middle: Interface layer for all other components
- Lowest: Cluster manager (Distributed computing)

Each functionalities:
- **Spark SQL:** A module for working with structured data, allowing users to query data using SQL syntax as well as Apache Hive queries.
- **Spark Streaming:** An extension of the core Spark API that enables scalable, high-throughput, fault-tolerant stream processing of live data streams.
- **MLlib:** The machine learning library providing a suite of scalable machine learning algorithms.
- **GraphX:** A library for manipulating graphs and performing graph-parallel computations.



In the next level of detail, the architecture of Spark includes:

- **Worker Nodes:** Each worker node in a Spark cluster contains executor processes to run computations and store data for the application.
- **Executors:** Executors are responsible for running the tasks assigned to them by the Spark Context and returning results to it. They also provide in-memory storage for RDDs that are cached by user programs.
- **Spark Context:** Acts as the "Task Master," creating a logical job that is then divided into tasks and distributed across the worker nodes.
- **Cluster Manager (only exist in the case of distributed computing):** An external service for acquiring resources on the cluster (e.g., Standalone Manager, YARN, or Mesos). It allocates resources to Spark applications based on the demands of each application.



While Spark abstracts much of the complexity in resource management and task distribution, it also offers low-level APIs for finer-grained control over these aspects. Here's what you need to know:

- **Low-Level APIs:** These include RDDs (Resilient Distributed Datasets) and accumulators. RDDs are the fundamental data structure of Spark that represents a collection of items distributed across many compute nodes that can be manipulated in parallel. Accumulators are variables that are only "added" to through an associative and commutative operation and are used to perform counters or sums.
- **Resource Management:** Spark can be directed to use certain resource allocation strategies, such as dynamic allocation, which allows Spark to dynamically adjust the number of executors based on the workload.

By using these low-level APIs, users can optimize their workflows by manually controlling partitions, parallelism, and other aspects of their Spark jobs.
