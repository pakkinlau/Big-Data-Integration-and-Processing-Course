Step 3: Specifying data folder location

If the previous step showing the OS can detect mongoDB. It still return some warning from mongoDB because some config is still missing. 

The next step is to define data folder.

Use the following command to setup data folder path: `mongod --dbpath "C:\path\to\your\data\directory"`. For me, it is `mongod --dbpath "C:\Users\kinla\Documents\mongoDB data folder"`

After running this command, a set of files will appear in the data folder. 

The terminal will stop sending out message at some point, it is a normal behavior of being a server. It something updated, then there will be new messages sending out. 



# Part 2: Python testing 

Install pymongo, and then run the following code: 

```python
from pymongo import MongoClient

# Connect to the MongoDB server running on localhost:27017 (default)
client = MongoClient('localhost', 27017)

# Specify the database and collection
db = client['test_database']
collection = db['test_collection']

# Create (Insert a document)
inserted_document = collection.insert_one({'name': 'Alice', 'email': 'alice@example.com'})
print(f'Inserted document id: {inserted_document.inserted_id}')

# Read (Query the database)
for document in collection.find():
    print(document)

# Update (Update a document)
collection.update_one({'name': 'Alice'}, {'$set': {'email': 'alice_updated@example.com'}})

# Verify the update
for document in collection.find():
    print(document)

# Delete (Remove a document)
collection.delete_one({'name': 'Alice'})

# Verify the deletion
for document in collection.find():
    print(document)
```
